```python
import pandas as pd
import numpy as np

def convert_score(score):
    if pd.isna(score) or score == '':
        return np.nan
    
    # Handle fraction formats (e.g., "3.5/4")
    if '/' in str(score):
        try:
            num, denom = map(float, score.split('/'))
            return num / denom
        except:
            return np.nan
    
    # Handle letter grades
    grade_map = {
        'A+': 0.95, 'A': 0.9, 'A-': 0.85,
        'B+': 0.8, 'B': 0.7, 'B-': 0.65,
        'C+': 0.6, 'C': 0.5, 'C-': 0.45,
        'D+': 0.4, 'D': 0.3, 'D-': 0.2,
        'F': 0.1
    }
    if str(score).upper() in grade_map:
        return grade_map[str(score).upper()]
    
    # Handle date-like strings (e.g., "5-Jan") - these are likely errors
    if '-' in str(score) and any(month in str(score).lower() for month in ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']):
        return np.nan
    
    # Try to convert remaining values to float
    try:
        return float(score)
    except:
        return np.nan

# Apply conversion
df['numeric_score'] = df['originalScore'].apply(convert_score)

# Drop rows where conversion failed
df = df.dropna(subset=['numeric_score'])
```

```python
import numpy as np

# Force scores to 0-1 range
df['numeric_score'] = np.clip(df['numeric_score'], 0, 1)

# Identify and remove remaining invalid conversions
invalid_mask = (df['numeric_score'] < 0) | (df['numeric_score'] > 1)
print(f"Found {invalid_mask.sum()} invalid scores to clean:")

# Show problematic rows
print(df.loc[invalid_mask, ['originalScore', 'numeric_score']].head(10))

# Drop invalid rows (recommended)
df = df[~invalid_mask]

### -- OR ---

## Option B: Or set them to NaN for manual review
# df.loc[invalid_mask, 'numeric_score'] = np.nan

# Verify cleanup
print("\nCleaned statistics:")
print(df['numeric_score'].describe())
```

```python
## dropping columns not needed for feature extraction

#reviewId -> Just an identifier, no impact on sentiment.

# creationDate -> Could be useful for time-based analysis, but may not directly influence sentiment.

# criticName -> Categorical; could be useful if specific critics are more negative/positive, but might not generalize well.

# publicatioName -> Similar to criticName, unless some publications are biased.

# reviewState -> Needs further analysis, but likely not critical.

df = df.drop(columns=['reviewId', 'creationDate', 'criticName', 'publicatioName', 'reviewState'])
```

```python
## extracting text based features from reviews
# longer reviews might show stronger opinions

df['word_count'] = df['reviewText'].apply(lambda x: len(str(x).split()))
df['char_count'] = df['reviewText'].apply(lambda x: len(str(x)))


df.head()
```

```python
def convert_fraction_to_float(score):
    try:
        if isinstance(score, str) and "/" in score:
            num, denom = score.split("/")  # Split "X/Y"
            return float(num) / float(denom)  # Convert to decimal
        else:
            return np.nan  # Handle missing values
    except:
        return np.nan  # Handle conversion errors

df['normalized_score'] = df['originalScore'].apply(convert_fraction_to_float)

df.head()
```

```python
### Feature encoding

## Encoding 'istopcritic' true/false val to 0/1 (unnecessary as it was already converted into bool during cleaning phase ) 
## And most ML model and panda automatically treats them as 0/1 of int type
df['isTopCritic'] = df['isTopCritic'].astype(int)


## Encoding year as ordinal feature
df['year'] = df['year'].astype(int)

df.head()
```

```python
# categorizing review based on length of 3-equal sized bins/categories
df['review_length_category'] = pd.qcut(df['review_length'], q=3, labels=['Short', 'Medium', 'Long'])

df.head(10)
```

```python
##  -----   Pearson Correlation (Numerical vs Numerical)   -------

# Pearson correlation measures linear relationships between numeric features (numeric_score, review_length, char_count, etc.)

# Select only numeric columns
numeric_cols = ['numeric_score', 'review_length', 'word_count', 'char_count', 'normalized_score']

# Compute correlation matrix
corr_matrix = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()


##      ------------    Interpretation      --------------

# High positive correlation (+1) means two features increase together.

# High negative correlation (-1) means one feature increases while the other decreases.

# Near 0 means little to no relationship
```

```python
##  -------------   Binary vs. Numeric Correlation (Point-Biserial) -----------------

# isTopCritic (0 or 1) and scoreSentiment (Positive/Negative) are binary

from scipy.stats import pointbiserialr

# Correlation between isTopCritic and numeric_score
r, p_value = pointbiserialr(df['isTopCritic'], df['numeric_score'])
print(f"Correlation (isTopCritic vs numeric_score): r = {r:.3f}, p = {p_value:.3f}")

# Correlation between isTopCritic and review_length
r, p_value = pointbiserialr(df['isTopCritic'], df['review_length'])
print(f"Correlation (isTopCritic vs review_length): r = {r:.3f}, p = {p_value:.3f}")


##  --------------- Interpretation  ------------------

# If r is positive, isTopCritic = 1 tends to have higher values in the numeric column.

# If r is negative, isTopCritic = 1 tends to have lower values.

# If p < 0.05, the correlation is statistically significant.
```
